## 1\. Preço do diamante por tipo de corte

````
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
````

````
# gráfico do exercício 1

diamonds = sns.barplot(x='cut', y='price', palette ='Blues', data = data)
diamonds.set_title('Preço do diamante por tipo de corte')
diamonds.set_xlabel('Tipo de corte')
diamonds.set_ylabel('Preço')
diamonds.set(ylim=(0, 5000))


````



```
# Isto está formatado como código
```

**Insight do gráfico 1**: ... # Os tipo de  corte parece ter uma relação com o valor do diamante, os diamantes com corte premium tem valores maiores nos preços seguidos do corte fair, já o corte ideal tem um valor menor comparado com os demais.

## 2\. Número de passageiros em dezembro por ano

````
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
````

````
# gráfico do exercício 2
flights = data.query('month == "Dec"')
sns.lineplot(x='year', y='passengers', data=flights)
sns.set_theme(style="darkgrid")




````

**Insight do gráfico 2**: ... # Podemos ver um consideravél aumento de passageiros do mês de dezembro de cada ano, nos anos 1957 e 1958 da pra ver que teve uma valor consideravél mente iguais de passageiros nesse mês, e nos proximos anos voltou a aumentar.

## 3\. Numero de passageiros por mês entre 1949 e 1959

````
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
````

````
# gráfico do exercício 3

flights = data.query('year >= 1949 and year <= 1959')
sns.lineplot(x='month', y='passengers', hue='year', data=flights)
sns.set_theme(style="darkgrid")
sns.set_context("paper", font_scale=1.5, rc={"lines.linewidth": 2.5})
sns.set_palette("Set2")

````

[texto do link](https://)**Insight do gráfico 3**: # Os meses de janeiro, fevereiro, e novembro mostra uma constante baixa de passageiros, já os meses junho, julho agosto e dezembro são os meses que mas tem uma constante alta de passageiros, em todos os anos mostrando uma mesma tendência.
